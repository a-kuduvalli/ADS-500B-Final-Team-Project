```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset
file_path = r"C:\Users\jrmoo\cleaned_house_sales.csv"
df = pd.read_csv(file_path)

# Select Numeric Columns
numeric_df = df.select_dtypes(include=[np.number])

# Correlation Matrix
corr_matrix = numeric_df.corr()

# Display Matrix
corr_matrix
```

```python
import numpy as np

# Dependent variable
dependent_variable = "price"

# Correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Correlations with the dependent variable
corr_with_price = corr_matrix[dependent_variable].drop(dependent_variable)

# Group predictors in 3 categories
strong_predictors = corr_with_price[abs(corr_with_price) > 0.50]
moderate_predictors = corr_with_price[(abs(corr_with_price) > 0.25) & (abs(corr_with_price) <= 0.50)]
weak_predictors = corr_with_price[abs(corr_with_price) <= 0.25]

# Display Results
print("Dependent Variable:")
print(dependent_variable)

print("\nIndependent Variables:")

print("\nStrong Predictors (0.50 ->):")
for col, val in strong_predictors.items():
    print(f"- {col} ({val:.4f})")

print("\nModerate Predictors (0.25 -> 0.50):")
for col, val in moderate_predictors.items():
    print(f"- {col} ({val:.4f})")

print("\nWeak Predictors (-> 0.25):")
for col, val in weak_predictors.items():
    print(f"- {col} ({val:.4f})")
```
