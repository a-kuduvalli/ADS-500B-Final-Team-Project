```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
file_path = r"C:\Users\jrmoo\cleaned_house_sales.csv"
df = pd.read_csv(file_path)

plt.style.use("ggplot")

# Select Numeric Columns
numeric_df = df.select_dtypes(include=[np.number])

# Correlation Matrix
corr_matrix = numeric_df.corr()

# Display Matrix
corr_matrix
```

Correlation Matrix:

The required libraries were imported, and the cleaned housing dataset was loaded into a pandas DataFrame. The code selects only the numerical variables in the dataset. Finally, a correlation matrix was generated to measure the strength and direction of relationships between numerical features. This matrix serves as the foundation for identifying strong, moderate, and weak predictors for later analysis.

```python
import numpy as np

# Dependent variable
dependent_variable = "price"

# Correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Correlations with the dependent variable
corr_with_price = corr_matrix[dependent_variable].drop(dependent_variable)

# Group predictors in 3 categories
strong_predictors = corr_with_price[abs(corr_with_price) > 0.50]
moderate_predictors = corr_with_price[(abs(corr_with_price) > 0.25) & (abs(corr_with_price) <= 0.50)]
weak_predictors = corr_with_price[abs(corr_with_price) <= 0.25]

# Display Results
print("Dependent Variable:")
print(dependent_variable)

print("\nIndependent Variables:")

print("\nStrong Predictors (0.50 ->):")
for col, val in strong_predictors.items():
    print(f"- {col} ({val:.4f})")

print("\nModerate Predictors (0.25 -> 0.50):")
for col, val in moderate_predictors.items():
    print(f"- {col} ({val:.4f})")

print("\nWeak Predictors (-> 0.25):")
for col, val in weak_predictors.items():
    print(f"- {col} ({val:.4f})")
```

Independent and Dependent Variables

This identifies how strongly each numerical variable is related to the target variable, price, by calculating their correlation values. After analyzing the correlation matrix, the code takes the correlation of every predictor with price and removes the dependent variable itself. Based on correlation strength, they are grouped into three categories:

•	Strong predictors (correlation ≥ 0.50)
•	Moderate predictors (0.25–0.50)
•	Weak predictors (≤ 0.25)

These groups help to determine which variables have the greatest influence on price and guide the selection of features for further exploratory analysis and modeling. Each group is then printed along with its correlation values for an easier understanding of our findings.

```python
# Exploratory Analysis with Visualizations

numeric_df = df.select_dtypes(include=[np.number])

corr_matrix = numeric_df.corr()
corr_matrix

dependent_variable = "price"

corr_with_price = corr_matrix[dependent_variable].drop(dependent_variable)

strong_predictors   = corr_with_price[abs(corr_with_price) >= 0.50]
moderate_predictors = corr_with_price[(abs(corr_with_price) > 0.25) & (abs(corr_with_price) < 0.50)]
weak_predictors     = corr_with_price[abs(corr_with_price) <= 0.25]

print("Strong predictors:\n", strong_predictors)
print("\nWeak predictors:\n", weak_predictors)

top3_strong = strong_predictors.sort_values(ascending=False).head(3).index.tolist()
top3_weak = weak_predictors.abs().sort_values().head(3).index.tolist()

print("Top 3 Strong Predictors:", top3_strong)
print("Top 3 Weak Predictors:", top3_weak)

sns.pairplot(df[top3_strong + ["price"]], diag_kind="kde", corner=True, plot_kws={'alpha':0.5})
plt.suptitle("Pairplot of Strongly Correlated Predictors", y=1.02)
plt.show()

sns.pairplot(df[top3_weak + ["price"]], diag_kind="kde", corner=True, plot_kws={'alpha':0.5})
plt.suptitle("Pairplot of Weakly Correlated Predictors", y=1.02)
plt.show()
```

Exploratory Visualization of Strong and Weak Predictors

The correlation analysis is extended by visualizing how the strongest and weakest predictors relate to the target variable, price. After identifying each feature’s correlation strength, the code selects the top three strongest predictors (sqft_living, grade, and sqft_above) and the top three weakest predictors (long, id, and condition). Pairplots are generated to display scatterplots and distribution curves for each predictor alongside the dependent variable.

The pairplot of strongly correlated predictors shows clear upward patterns, especially for living area and above-ground square footage, confirming their strong positive influence on home prices. The structure of these plots shows that as these variables increase, the price tends to increase as well.

To show contrast between strong and weak predictors, I also created visualizations of the 3 weakest variables in relation to our target variable. The pairplot of weakly correlated predictors shows no meaningful patterns between these variables and price. The scatterplots appear widely dispersed, and the distributions show little to no relationship, indicating that these features contribute minimally to predicting home value.

These visualizations help to validate the numerical correlation results and show which features are most suitable for building accurate predictive models.

Distribution Analysis for Key Numerical Features

This section visualizes the distribution of important numerical features in the dataset: price, sqft_living, sqft_lot, bathrooms, and bedrooms—using histograms. These plots help reveal the overall shape, skewness, and spread of each variable.

The distributions show that price is heavily right-skewed, meaning most homes fall within lower price ranges, with fewer high-priced outliers. Sqft_living follows a similar pattern, where most homes have moderate living space and only a small number have very large square footage. Sqft_lot displays extreme right skew, indicating that a few homes sit on unusually large lots compared to the majority. The bathrooms and bedrooms features also show distinct clustering around typical home layouts with 2–3 bedrooms and 1.5–2.5 bathrooms, with fewer homes having unusually high counts.

These visuals provide a clearer picture of how each feature is distributed, and help identify skewed variables and potential outliers that may impact modeling.


```python
sns.lmplot(
    data=df,
    x='sqft_living',
    y='price',
    scatter_kws={'alpha': 0.3, 'color': 'salmon'},
    line_kws={'color': 'blue', 'linewidth': 2}
)

plt.title("Price vs Sqft Living (With Regression Line)")
plt.show()
```

Price vs. Sqft Living

This scatterplot illustrates the relationship between sqft_living and price, with each point representing a home. The regression line is drawn in blue to clearly stand out from the data points, making the trend easier to interpret.

The plot shows a strong positive linear relationship, meaning that as the interior living space increases, housing prices tend to rise as well. This matches the high correlation value previously calculated (≈ 0.68). The upward-sloping blue regression line reinforces this trend and highlights the overall direction of the relationship despite natural variation within the data.

A cluster of high-sqft or high-price outliers appears above the main cluster, suggesting that specific homes with very large living areas behave differently from the rest of the homes.
